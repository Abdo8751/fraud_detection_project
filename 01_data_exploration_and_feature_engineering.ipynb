**Reasoning**:
Based on the identified missing values, I will handle them by imputing `DeductibleAmtPaid` with 0, and filling the missing physician, diagnosis, and procedure codes with 'Missing' to indicate their absence. This strategy is chosen because 0 is a sensible default for an amount not paid, and 'Missing' is appropriate for categorical codes to avoid loss of information or misinterpretation.



```python
final_merged_df['DeductibleAmtPaid'] = final_merged_df['DeductibleAmtPaid'].fillna(0)

object_cols_to_fill = [
    'AttendingPhysician', 'OperatingPhysician', 'OtherPhysician',
    'ClmAdmitDiagnosisCode', 'DiagnosisGroupCode',
    'ClmDiagnosisCode_1', 'ClmDiagnosisCode_2', 'ClmDiagnosisCode_3', 'ClmDiagnosisCode_4',
    'ClmDiagnosisCode_5', 'ClmDiagnosisCode_6', 'ClmDiagnosisCode_7', 'ClmDiagnosisCode_8',
    'ClmDiagnosisCode_9', 'ClmDiagnosisCode_10'
]

for col in object_cols_to_fill:
    final_merged_df[col] = final_merged_df[col].fillna('Missing')

# Procedure codes are float64 and have many NaNs. Convert to object then fill with 'Missing'.
procedure_cols = [
    'ClmProcedureCode_1', 'ClmProcedureCode_2', 'ClmProcedureCode_3',
    'ClmProcedureCode_4', 'ClmProcedureCode_5', 'ClmProcedureCode_6'
]
for col in procedure_cols:
    final_merged_df[col] = final_merged_df[col].astype(str).replace({'nan': 'Missing'})

print("Missing values handled for specified columns.")
print(final_merged_df.isnull().sum()[final_merged_df.isnull().sum() > 0])
```

## Data Cleaning and Preprocessing

### Subtask:
Continue handling missing values and correcting data types, specifically for `RenalDiseaseIndicator` and `PotentialFraud` columns, and check for duplicate rows in `final_merged_df`.


**Reasoning**:
I will convert the 'RenalDiseaseIndicator' column to numeric, mapping '0' to 0 and '1' to 1, handling any non-numeric values by coercing them to NaN and then filling with 0. After that, I will convert 'PotentialFraud' from 'Yes'/'No' to 1/0, and finally identify and remove duplicate rows to ensure data quality and uniqueness.



```python
import numpy as np

# 1. Convert 'RenalDiseaseIndicator' to integer type
# The column contains '0' and '1' as strings. '2' is also present in some chronic condition columns, but not in RenalDiseaseIndicator based on initial head, still safer to coerce.
final_merged_df['RenalDiseaseIndicator'] = pd.to_numeric(final_merged_df['RenalDiseaseIndicator'], errors='coerce').fillna(0).astype(int)

# 2. Convert 'PotentialFraud' to numerical type (1 for 'Yes', 0 for 'No')
final_merged_df['PotentialFraud'] = final_merged_df['PotentialFraud'].map({'Yes': 1, 'No': 0})

print("Data types for 'RenalDiseaseIndicator' and 'PotentialFraud' updated successfully.")

# 3. Check for and remove duplicate rows
print(f"Shape of DataFrame before removing duplicates: {final_merged_df.shape}")
duplicate_rows = final_merged_df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_rows}")

if duplicate_rows > 0:
    final_merged_df.drop_duplicates(inplace=True)
    print(f"Shape of DataFrame after removing duplicates: {final_merged_df.shape}")
else:
    print("No duplicate rows to remove.")

print("First 5 rows of the DataFrame after cleaning and handling duplicates:")
final_merged_df.head()
```

## Aggregate Data by Provider

### Subtask:
Aggregate the cleaned data by `Provider` to summarize financial and procedural claim information.


**Reasoning**:
First, I will calculate the `ClaimDuration` by subtracting `ClaimStartDt` from `ClaimEndDt` and converting the result to days, as this is a prerequisite for aggregation. Then, I will define the columns for diagnosis and procedure codes to prepare for counting unique non-missing values. Finally, I will group the data by `Provider` and `PotentialFraud` and calculate all the specified aggregate statistics, including the unique diagnosis and procedure codes, as described in the instructions.



```python
final_merged_df['ClaimDuration'] = (final_merged_df['ClaimEndDt'] - final_merged_df['ClaimStartDt']).dt.days

# Define diagnosis and procedure code columns
diagnosis_cols = [f'ClmDiagnosisCode_{i}' for i in range(1, 11)]
procedure_cols = [f'ClmProcedureCode_{i}' for i in range(1, 7)]

# Aggregate the data by Provider and PotentialFraud
provider_aggregated_df = final_merged_df.groupby(['Provider', 'PotentialFraud']).agg(
    TotalClaimAmtReimbursed=('InscClaimAmtReimbursed', 'sum'),
    TotalDeductibleAmtPaid=('DeductibleAmtPaid', 'sum'),
    TotalIPAnnualReimbursementAmt=('IPAnnualReimbursementAmt', 'sum'),
    TotalOPAnnualReimbursementAmt=('OPAnnualReimbursementAmt', 'sum'),
    NumClaims=('ClaimID', 'count'),
    NumUniqueBeneficiaries=('BeneID', lambda x: x.nunique()),
    AvgClaimDuration=('ClaimDuration', 'mean'),
    NumUniqueDiagnosisCodes=('ClaimID', lambda x: final_merged_df.loc[x.index, diagnosis_cols].stack().loc[lambda s: s != 'Missing'].nunique()),
    NumUniqueProcedureCodes=('ClaimID', lambda x: final_merged_df.loc[x.index, procedure_cols].stack().loc[lambda s: s != 'Missing'].nunique())
).reset_index()

print("Aggregated data by Provider and PotentialFraud successfully.")
print("First 5 rows of provider_aggregated_df:")
print(provider_aggregated_df.head())
print("\nInformation about provider_aggregated_df:")
provider_aggregated_df.info()
```

# 9. Outlier Detection and Winsorization

##Subtask:
## Identify outliers using IQR method and cap extreme values at 1st and 99th percentiles.
### Reasoning:
Outliers can distort model training:
- IQR method identifies extreme values.
- Winsorization caps extreme values while preserving rank order.
- Log transforms and winsorized features together improve stability for modeling.

```python

#from scipy.stats.mstats import winsorize

monetary_cols = [c for c in num_cols if 'Total' in c or 'Reimbursement' in c or 'ClaimAmt' in c]
monetary_cols
# IQR-based flagging
outlier_flags = {}
for c in monetary_cols:
    Q1 = df[c].quantile(0.25)
    Q3 = df[c].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
    outlier_flags[c] = df[(df[c] < low) | (df[c] > high)].shape[0]
pd.Series(outlier_flags).sort_values(ascending=False)
# Optional: create winsorized columns (cap at 1st and 99th percentile)
for c in monetary_cols:
    lower = df[c].quantile(0.01)
    upper = df[c].quantile(0.99)
    df[f"{c}_winsor"] = np.clip(df[c], lower, upper)
```

# 10. Feature Engineering
### Reasoning:
Feature engineering enhances model predictive power:
- Ratios capture normalized metrics (per claim, per beneficiary) that reduce scale effects.
- Interaction features capture multiplicative effects between claims and amounts.
- Binary flags highlight extreme behaviors (e.g., high claim amounts, many procedures).
- Log transformations reduce skewness and improve numerical stability.
- All engineered features are intended to help detect fraudulent patterns more clearly.

```python
# Safe division utility
def safe_div(a,b):
    return a / b.replace(0, np.nan)

# 1) Avg claim amount per claim (Total / NumClaims)
df['avg_claim_amt_per_claim'] = safe_div(df['TotalClaimAmtReimbursed'], df['NumClaims']).fillna(0)

# 2) avg claim amt per beneficiary
df['avg_claim_amt_per_beneficiary'] = safe_div(df['TotalClaimAmtReimbursed'], df['NumUniqueBeneficiaries']).fillna(0)

# 3) beneficiaries per claim
df['beneficiaries_per_claim'] = safe_div(df['NumUniqueBeneficiaries'], df['NumClaims']).fillna(0)

# 4) IP vs OP ratio (IP / OP)
df['ip_op_ratio'] = safe_div(df['TotalIPAnnualReimbursementAmt'], df['TotalOPAnnualReimbursementAmt']).replace([np.inf, -np.inf], np.nan).fillna(0)

# 5) high_claim_share (claims above some large-dollar threshold per provider) - approximate using winsorized amounts
threshold = df['TotalClaimAmtReimbursed'].quantile(0.95)
df['high_total_claim_flag'] = (df['TotalClaimAmtReimbursed'] > threshold).astype(int)

# 6) claim intensity: claims per active day (if AvgClaimDuration approximates activity)
df['claims_per_day'] = safe_div(df['NumClaims'], df['AvgClaimDuration']).replace([np.inf,-np.inf], np.nan).fillna(0)

# 7) diversity features already present: NumUniqueDiagnosisCodes, NumUniqueProcedureCodes
# 8) ratios of unique codes per claim
df['diag_per_claim'] = safe_div(df['NumUniqueDiagnosisCodes'], df['NumClaims']).fillna(0)
df['proc_per_claim'] = safe_div(df['NumUniqueProcedureCodes'], df['NumClaims']).fillna(0)

# 9) percent inpatient claims (approx) if we have IP/OP counts; if not, use ip_op_ratio as proxy
# (we'll keep ip_op_ratio)

# 10) log transforms for skewed numeric variables (create new columns)
to_log = ['TotalClaimAmtReimbursed','TotalIPAnnualReimbursementAmt','TotalOPAnnualReimbursementAmt',
          'NumClaims','NumUniqueBeneficiaries','avg_claim_amt_per_claim','avg_claim_amt_per_beneficiary']
for c in to_log:
    df[f"{c}_log1p"] = np.log1p(df[c].clip(lower=0))  # clip negative if any

# 11) interaction terms: e.g., (NumClaims * avg_claim_amt_per_claim)
df['claims_x_avgclaim'] = df['NumClaims'] * df['avg_claim_amt_per_claim']

# 12) indicator for many unique procedures (possible upcoding)
proc_thresh = df['NumUniqueProcedureCodes'].quantile(0.90)
df['many_procs_flag'] = (df['NumUniqueProcedureCodes'] > proc_thresh).astype(int)

# Log transforms for skewed numeric variables
to_log = ['TotalClaimAmtReimbursed','TotalIPAnnualReimbursementAmt','TotalOPAnnualReimbursementAmt',
          'NumClaims','NumUniqueBeneficiaries','avg_claim_amt_per_claim','avg_claim_amt_per_beneficiary']
for c in to_log:
    df[f"{c}_log1p"] = np.log1p(df[c].clip(lower=0))
```
